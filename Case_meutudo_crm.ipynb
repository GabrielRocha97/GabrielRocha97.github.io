3. Na aba “base_canais“ temos os dados de cada um dos canais utilizados ao longo do período
de duração deste teste AB. Sumarize esses dados por canal e monte o funil descrito acima
(com as conversões entre cada uma das 4 etapas).


R.:

```python
import pandas as pd
import numpy as np
```

```python
path = "/content/Case de CRM_ dados.xlsx"
```

```python
base_canais = pd.read_excel(path, sheet_name="base_canais")

base_canais.head()
```

```python
base_canais.info()
base_canais.describe()
```

```python
agg = (
    base_canais
    .groupby("canal", as_index=False)
    .agg(
        audiencia_total=("audiência", "sum"),
        envios_total=("envios", "sum"),
        clicks_total=("clicks", "sum"),
        conversoes_total=("conversões", "sum"),
    )
)

agg
```

```python
funil = agg.copy()

funil["taxa_alcance"] = funil["envios_total"] / funil["audiencia_total"]
funil["taxa_engajamento"] = funil["clicks_total"] / funil["envios_total"]
funil["taxa_conv_click"] = funil["conversoes_total"] / funil["clicks_total"]
funil["taxa_conv_final"] = funil["conversoes_total"] / funil["audiencia_total"]

funil = funil.round(4)
funil
```

5. Usando os dados de produção e custo por canal, também disponíveis na aba “base_canais“,
estime o quanto de lucro cada um dos canais de comunicação gerou. Assuma que nós
recebemos como receita 5% do total do valor da produção gerada.

R.:

```python
path = "/content/Case de CRM_ dados.xlsx"
base_canais = pd.read_excel(path, sheet_name="base_canais")
```

```python
agg_lucro = (
    base_canais
    .groupby("canal", as_index=False)
    .agg(
        producao_total=("produção", "sum"),
        custo_total=("custo", "sum")
    )
)

agg_lucro
```

```python
agg_lucro["receita_5pct"] = agg_lucro["producao_total"] * 0.05
agg_lucro["lucro"] = agg_lucro["receita_5pct"] - agg_lucro["custo_total"]

agg_lucro = agg_lucro.round(2)
agg_lucro
```

7. Alguns dos usuários impactados por uma jornada de CRM, converteriam mesmo se não
houvessem sido impactados por essa jornada. Assim, para estimar de forma mais precisa o lift
de conversões e produção podemos criar um grupo de controle no teste AB (grupo de usuários
que que não sofre impacto de nenhuma das jornadas). Na aba “base_grupo_controle” da
mesma sheets, há dados de um grupo de controle que rodou no mesmo teste AB dos canais.
Utilize esses dados para re-estimar o lucro de cada um dos canais.

R.:

```python
path = "/content/Case de CRM_ dados.xlsx"
base_controle = pd.read_excel(path, sheet_name="base_grupo_controle")

aud_controle = base_controle["audiência"].sum()
conv_controle = base_controle["conversões"].sum()
prod_controle = base_controle["produção"].sum()

##receita natural do grupo controle (5%)##
receita_controle_5pct = prod_controle * 0.05

print("== GRUPO DE CONTROLE ==")
print("Audiência total:", aud_controle)
print("Conversões totais:", conv_controle)
print("Produção total (R$):", round(prod_controle, 2))
print("Receita natural (5%):", round(receita_controle_5pct, 2))

#taxa base e ticket base#
taxa_conv_base = conv_controle / aud_controle
ticket_medio_base = prod_controle / conv_controle

print("\nTaxa base de conversão:", round(taxa_conv_base, 6))
print("Ticket médio do controle (R$):", round(ticket_medio_base, 2))

base_canais = pd.read_excel(path, sheet_name="base_canais")
agg = (
    base_canais
    .groupby("canal", as_index=False)
    .agg(
        audiencia_total=("audiência", "sum"),
        conversoes_total=("conversões","sum"),
        producao_total=("produção","sum"),
        custo_total=("custo","sum")
    )
)

##baseline (sem CRM##
agg["conv_esp_controle"] = agg["audiencia_total"] * taxa_conv_base
agg["prod_esp_controle"] = agg["conv_esp_controle"] * ticket_medio_base
agg["receita_esp_controle_5pct"] = agg["prod_esp_controle"] * 0.05   # << ADICIONADO

##incrementos (efeito do CRM)##
agg["prod_incremental"] = agg["producao_total"] - agg["prod_esp_controle"]
agg["receita_incremental_5pct"] = agg["prod_incremental"] * 0.05
agg["lucro_incremental"] = agg["receita_incremental_5pct"] - agg["custo_total"]

##mostrar resultado final##
print("\n=== RESULTADO FINAL — LUCRO INCREMENTAL POR CANAL ===")
agg[[
    "canal",
    "producao_total",
    "prod_esp_controle",
    "prod_incremental",
    "receita_esp_controle_5pct",      # << ADICIONADO (5% baseline)
    "receita_incremental_5pct",
    "custo_total",
    "lucro_incremental"
]].round(2)
```
